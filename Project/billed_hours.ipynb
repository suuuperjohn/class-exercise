```python
import pandas as pd
gdp=pd.read_excel('GDP_su.xlsx', sheet_name='Sheet2')
#barpass= pd.read_excel('Bar Pass Stats_su.xlsx')
#grad=pd.read_excel('Gradstudents_su.xlsx', sheet_name='Sheet2')
perincome=pd.read_excel('Percapital_income_su.xlsx', sheet_name='Sheet2')
perdispose=pd.read_excel('personal_disposable_income_su.xlsx', sheet_name='Sheet2')
population=pd.read_excel('population2014_2022.xlsx', sheet_name='Sheet2')
employ=pd.read_excel('Total_employment_su.xlsx', sheet_name='Sheet2')
bankruptcy=pd.read_excel('Totals_su.xlsx', sheet_name='Sheet1')
billed_hours=pd.read_excel('9_year_worked_and_billed_hours_ver2.xlsx', sheet_name="Sheet2")
```

```python
#df= pd.merge(left=gdp, right=barpass,how='outer', on=['Year','Jurisdiction'])
#df= pd.merge(left=df, right=grad,how='outer', on=['Year','Jurisdiction'])
df= pd.merge(left=gdp, right=perincome,how='outer', on=['Year','Jurisdiction'])
df= pd.merge(left=df, right=perdispose,how='outer', on=['Year','Jurisdiction'])
df= pd.merge(left=df, right=population,how='outer', on=['Year','Jurisdiction'])
df= pd.merge(left=df, right=employ,how='outer', on=['Year','Jurisdiction'])
df= pd.merge(left=df, right=bankruptcy,how='outer', on=['Year','Jurisdiction'])
df= pd.merge(left=df, right=billed_hours,how='outer', on=['Year','Jurisdiction'])
```

```python
df=df[(df.Year>2013)&(df.Year<2023)&(df.Jurisdiction!='US')]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipeline = Pipeline([
    ('imp', SimpleImputer(strategy='mean')),##dealing with nas with median function.
    ('norm', StandardScaler()),##scaling 
])##like a function that use to deal with your df, we have not apply it to data yet.
```

```python
col_transformer= ColumnTransformer([
    ('num', num_pipeline,['GDP', 'Percapital_income', 'personal_disposable_income',
       'population', 'total_emplyment', 'Bankruptcy', 'month',
       'total_worked_hours']),
    ('cato', OneHotEncoder(),['office_location','department'])
])
```

```python
df.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
full_pipeline= Pipeline([
    ('trans', col_transformer),
    ('cls', LinearRegression())
])
```

```python
df_X= df.drop(['Year','total_billed_hours'], axis= 'columns')
df_y= df.total_billed_hours
```

```python
X_train, X_test, y_train, y_test=train_test_split(df_X
                                                , df_y
                                                , test_size=0.3
                                                , stratify=df.Year)
```

```python
X_train.head(10)
```

```python
full_pipeline.fit(X_train, y_train)
```

```python
pickle.dump(full_pipeline, open('pipeline_linear_billed_hours.pkl', 'wb'))
##save your model as a pkl file
```
